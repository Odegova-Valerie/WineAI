```python
def strength_model():
    strength_model = joblib.load('strength.pkl')
    return strength_model
```

```python
def sugar_model():
    sugar_model = joblib.load('sugar.pkl')
    return sugar_model
```

```python
# Функция оценки приспособленности
strength = strength_model()
sugar = sugar_model()

def evaluate_blend(blend, target_strength, target_sugar):
    pred_strength = strength.predict(blend.reshape(-1, 48))
    pred_sugar = sugar.predict(blend.reshape(-1, 48))
    fitness = 1 / ((pred_strength - target_strength)**2 + (pred_sugar - target_sugar)**2)
    return fitness, pred_strength, pred_sugar
```

```python
def selection(population, fitness_scores):
    fitness_scores = np.array(fitness_scores)
    sorted_indices = np.argsort(-fitness_scores)
    return population[sorted_indices[:int(population_size/2)]]
```

```python
def crossover(parents, offspring_size, keep_parent_percent_index):
    offspring = []
    while len(offspring) < offspring_size:
        idx1, idx2 = np.random.choice(len(parents), 2, replace=False)
        parent1, parent2 = parents[idx1], parents[idx2]
   
        child = np.empty_like(parent1)

        child[:11] = parent1[:11]
        child[11:22] = parent1[11:22]

        parent_percent_index = [slice(22, 33), slice(33, 44)]

        child[22:33] = parent2[22:33]
        child[33:44] = parent2[33:44]

        child[parent_percent_index[keep_parent_percent_index]] = parent1[parent_percent_index[keep_parent_percent_index]]

        child[44:49] = parent2[44:49]
    
        offspring.append(child)

        
    return np.array(offspring)

def mutation(offspring, mutation_rate):
    if np.random.rand() < mutation_rate:
        row_index = np.random.randint(offspring.shape[0])
        percent_index = np.random.choice([10, 21, 32, 43])
        offspring[row_index, percent_index] += np.random.uniform(-0.1, 0.1)
        offspring[row_index] /= np.sum(offspring[row_index])

   
    return offspring
```

```python
# Обновление популяции новыми особями
def create_new_population(parents, offspring):
    new_population = np.vstack((parents, offspring))
    return new_population
```

```python
# Основной цикл генетического алгоритма
target_strength = 3
target_sugar = 2
mutation_rate = 0.1
generations = 50
keep_parent_percent_index = 0


for generation in range(generations):
    print('Поколение:', generation)

    fitness_scores = np.array([evaluate_blend(elem, target_strength, target_sugar)[0] for elem in population])
    fitness_scores = fitness_scores.flatten()
    
    best_idx = np.argmax(fitness_scores)

    best_blend = population[best_idx]
    best_blend = population[best_idx]
    best_fitness = fitness_scores[best_idx]

    best_fitness, best_strength, best_sugar = evaluate_blend(best_blend, target_strength, target_sugar)

    best_fitness = max(best_fitness)

    print(best_fitness, best_strength, best_sugar)
    
    if best_fitness >= 2:
        print(f"Найден подходящий купаж на {generation} поколении с крепостью {best_strength} и с содержанием сахара {best_sugar}.")
        break
    
    parents = selection(population, fitness_scores)
    offspring_size = population_size - len(parents)
    offspring_crossover = crossover(parents, offspring_size, keep_parent_percent_index)
    offspring_mutation = mutation(offspring_crossover, mutation_rate)
    population = create_new_population(parents, offspring_mutation)
```
