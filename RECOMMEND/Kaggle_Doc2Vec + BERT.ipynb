```python
#Import all the dependencies
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
from nltk.tokenize import word_tokenize
import nltk
nltk.download('punkt')
```

```python
data = new_df['description'].to_list()
tagged_data = [TaggedDocument(words=word_tokenize(_d.lower()), tags=[str(i)]) for i, _d in enumerate(data)]
```

```python
max_epochs = 100
vec_size = 1000
alpha = 0.025

model = Doc2Vec(vector_size=vec_size,
                alpha=alpha, 
                min_alpha=0.00025,
                min_count=1,
                dm=1)
  
model.build_vocab(tagged_data)

for epoch in range(max_epochs):
    print('iteration {0}'.format(epoch))
    model.train(tagged_data,
                total_examples=model.corpus_count,
                epochs=model.epochs)
    # decrease the learning rate
    model.alpha -= 0.0002
    # fix the learning rate, no decay
    model.min_alpha = model.alpha

model.save("d2v.model")
print("Model Saved")
```

```python
import torch
from transformers import AutoTokenizer, AutoModel
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import normalize

# Загрузка предобученного BERT
model_name = "bert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)


data = new_df['description'].to_list() 
user_input = "I want a dry red wine with citrus flavors. I have baked beef for dinner tonight. But the wine should be sour"  # Замените на запрос пользователя

# Токенизация и кодирование описаний вина
encoded_data = tokenizer(data, padding=True, truncation=True, return_tensors='pt')

# Получение векторов из BERT
with torch.no_grad():
    outputs = model(**encoded_data)
    embeddings = outputs.last_hidden_state[:, 0, :].numpy()

# Нормализация векторов
normalized_embeddings = normalize(embeddings)

# Токенизация и кодирование введенного пользователем запроса
encoded_user_input = tokenizer(user_input, return_tensors='pt')

# Получение вектора из BERT для запроса пользователя
with torch.no_grad():
    user_input_embedding = model(**encoded_user_input).last_hidden_state[:, 0, :].numpy()

# Нормализация вектора запроса пользователя
normalized_user_input_embedding = normalize(user_input_embedding)

# Рассчитываем косинусную близость между векторами
cosine_similarities = cosine_similarity(normalized_user_input_embedding, normalized_embeddings)

# Находим индекс наиболее близкого вектора
most_similar_index = cosine_similarities.argmax()

# Выводим результат
print(f"Наиболее близкий отзыв: {data[most_similar_index]}")
```
